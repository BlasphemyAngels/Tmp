# Pandas透视表与交叉表

```python
import pandas as pd
import numpy as np
```

```python
dataset_path = './starcraft.csv'
df_data = pd.read_csv(dataset_path, usecols=['LeagueIndex', 'Age', 'HoursPerWeek', 
                                             'TotalHours', 'APM'])
```

* 透视表 （pivot table）

```python
pd.pivot_table(df_data, 
               index=['GameID'],
               columns=['LeagueIndex'],
               aggfunc=np.sum)
```
