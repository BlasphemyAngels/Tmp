# Pandas透视表与交叉表

```python
import pandas as pd
import numpy as np
```

```python
dataset_path = './starcraft.csv'
df_data = pd.read_csv(dataset_path, usecols=['LeagueIndex', 'Age', 'HoursPerWeek', 
                                             'TotalHours', 'APM'])
```

* 透视表 （pivot table）

```python
pd.pivot_table(df_data, 
               index=['Age'],
               columns=['LeagueIndex'],
               aggfunc=np.sum)
```

```python
pd.pivot_table(df_data, 
               index=['Age'],
               columns=['LeagueIndex'],
               aggfunc=np.sum,
               fill_value=-100.)
```

* 交叉表 (crosstab)

```python
pd.crosstab(df_data['LeagueIndex'],
            df_data['Age'])
```

```python
pd.crosstab(df_data['LeagueIndex'],
            [df_data['Age'], df_data['HoursPerWeek']])
```
